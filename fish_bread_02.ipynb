```python
class BungeoppangShop:
    def __init__(self):
        self.stock = {
            '팥붕어빵':10,
            '슈크림붕어빵':8,
            '초코붕어빵':5
        }
        self.prices = {
            '팥붕어빵':1000,
            '슈크림붕어빵':1200,
            '초코붕어빵':1500
        }
        self.sales = {
            '팥붕어빵':0,
            '슈크림붕어빵':0,
            '초코붕어빵':0
        }

    def check_stock(self):
        print('현재 붕어빵 재고')
        for bread, count in self.stock.items():
            print(f'{bread}:{count}개')

    def process_order(self, bread_type, bread_count):
        try:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
            else :
                print('죄송합니다. 현재 재고가 부족합니다.')
        except:
            print('없는 메뉴입니다.')

    def admin_mode(self):
        while True:
            bread_type = input('추가할 붕어빵(팥붕어빵, 슈크림붕어빵, 초코붕어빵)을 입력해 주세요. 종료를 입력하시면 추가가 종료됩니다.:' )
            if bread_type == '종료':
                print('추가를 종료합니다.')
                break
            try:
                additional_stock = int(input('추가할 갯수를 입력하세요:'))
                self.stock[bread_type] += additional_stock
                print(f'{bread_type}이 {additional_stock}개 추가되었습니다.')
                self.check_stock()
            except:
                print('존재하지 않는 메뉴입니다.')

    def calculate_total_sales(self):
        total_sales = sum(self.sales[bread] * self.prices[bread] for bread in self.sales)
        print(f'총 가격: {total_sales}원')

def main():
    shop = BungeoppangShop()
    while True:
        mode = input('원하는 모드를 선택해 주세요.(주문, 관리자, 종료): ')
        if mode == '종료':
            break
        if mode == '관리자':
            shop.admin_mode()
            continue
        if mode == '주문':
            bread_type = input('추가할 붕어빵(팥붕어빵, 슈크림붕어빵, 초코붕어빵)을 입력해 주세요. 종료를 입력하시면 추가가 종료됩니다.:')
            if bread_type == '종료':
                break
            try:
                bread_count = int(input('주문할 빵의 갯수를 입력해주세요!: '))
                shop.process_order(bread_type, bread_count)
            except:
                print('정수로 입력해주세요!')
            shop.check_stock()
    shop.calculate_total_sales()

if __name__ == '__main__':
    main()


```
